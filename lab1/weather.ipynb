```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
```

```python
weather = pd.read_csv('weather2019.csv')
```

```python
weather.head()
```

```python
#1 How much rain fell on 17-Mar-19 (to 1 decimal place)?
rain = weather[weather['date']=='17-Mar-19']
print(rain.head())
```

```python
#2 Find the mean hours of sunlight per day (round your answer to 2 decimal places).
sunMean = weather['sun'].mean()
print("%.2f" %sunMean)
```

```python
#3 What was the lowest maximum daily temperature (round your answer to 1 decimal places)?
lowMaxTemp = weather['maxtemp'].min()
print(lowMaxTemp)
```

```python
#4 Find the median of the minimum daily temperature (round your answer to 1 decimal places).
medMinDailyTemp = weather['mintemp'].median()
print(medMinDailyTemp)
```

```python
#5 Find the standard deviation of the daily windspeed (round your answer to 2 decimal places).
stdDailyWindSpeed = weather['wind'].std()
print('The standard deviation of the daily wind speed', stdDailyWindSpeed,'%.')
```

```python
#6 On what date was the lowest minimum temperature recorded (state your answer in day-month-year format, e.g. 21-Apr-97)?
lowMinTemp = weather[weather['mintemp'] == weather['mintemp'].min()]
print(lowMinTemp.head())
```

```python
#7 Create a scatter plot of the maximum daily temperature against the daily rainfall
plt.scatter(weather.maxtemp,weather.rain)
```

```python
#8 Create a histogram for the average wind speed. Which of the following best describes the plot?
plt.hist(weather.wind,alpha=0.5, label="wind")
```
